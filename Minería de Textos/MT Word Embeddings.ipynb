```python
import gensim, time, re, sklearn
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from gensim.models import KeyedVectors, Word2Vec
from gensim.test.utils import datapath
```

```python
def import_and_preproc(path):
    """
    Abre y preprocesa el archivo de texto en cuestión.

    path = str ; Dirección del archivo.
    """
    aux = open(path, 'r')
    aux_opened = aux.readlines()
    final = [re.sub('[\n]', '', _) for _ in aux_opened]
    final = [_.split() for _ in final]
    return final

def single_list(word_list):
    """
    Lidia con "palabras compuestas".

    word_list = list ; Obtenida de la función import_and_preproc()
    """
    aux = []
    for _ in word_list:
        for j in _:
            aux.append(j)
    return aux
```

```python
food = single_list(import_and_preproc(r'C:\Users\FLopezP\Documents\GitHub\PCIC\Minería de Textos\Archivos\food_words.txt'))
sports = single_list(import_and_preproc(r'C:\Users\FLopezP\Documents\GitHub\PCIC\Minería de Textos\Archivos\sports_words.txt'))
weather =  single_list(import_and_preproc(r'C:\Users\FLopezP\Documents\GitHub\PCIC\Minería de Textos\Archivos\weather_words.txt'))
print(food[:5], sports[:5], weather[:5])
```

```python
start_time = time.time()
w2v = KeyedVectors.load_word2vec_format(datapath(r"GoogleNews-vectors-negative300.bin.gz"), binary=True)
fastText = KeyedVectors.load_word2vec_format(datapath("wiki-news-300d-1M-subword.vec"), binary = False)
glove = KeyedVectors.load_word2vec_format(datapath('glove.6B.300d.txt'), binary = False, no_header = True)
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
def get_vectors(lista, emb_model):
    """
    lista = list ; Proveniente de la función single_list()
    emb_model = str ; 'w2v', 'fastT', 'glove' Determina el modelo de embeddings.
    """
    if emb_model == 'w2v':
        model = w2v
    elif emb_model == 'fastT':
        model = fastText
    elif emb_model == 'glove':
        model = glove
    else:
        print('No es un modelo válido')

    
    vec_list = []
    none_vec = []
    for _ in lista:
        try:
            vector = model.get_vector(_)
            vec_list.append(vector)
        except KeyError:
            none_vec.append(_)
            
    return np.array(vec_list), none_vec
```

```python
def full_pipe(emb_model):
    """
    Genera una comparación entre las 3 
    
    emb_model = str ;  'w2v', 'fastT', 'glove'
    """
    food_vecs, food_none = get_vectors(food, emb_model)
    sports_vecs, sports_none = get_vectors(sports, emb_model)
    weather_vecs, weather_none = get_vectors(weather, emb_model)

    print("Palabras no encontradas")
    print(f"Food: {food_none}")
    print("---")
    print(f"Sports: {sports_none}")
    print("---")
    print(f"Weather: {weather_none}")
    print("---")

    whole_embs = np.concatenate((np.concatenate((food_vecs, sports_vecs)), weather_vecs))
    tsne = TSNE(n_components = 2, learning_rate = 'auto', init = 'random', perplexity = 10).fit_transform(whole_embs)

    food_x = [_[0] for _ in tsne[:len(food_vecs)]]
    food_y = [_[1] for _ in tsne[:len(food_vecs)]]

    sports_x = [_[0] for _ in tsne[len(food_vecs):(len(food_vecs) + len(sports_vecs))]]
    sports_y = [_[1] for _ in tsne[len(food_vecs):(len(food_vecs) + len(sports_vecs))]]

    weather_x = [_[0] for _ in tsne[(len(food_vecs) + len(sports_vecs)):]]
    weather_y = [_[1] for _ in tsne[(len(food_vecs) + len(sports_vecs)):]]

    food_vecs = np.array([food_x, food_y])
    sports_vecs = np.array([sports_x, sports_y])
    weather_vecs = np.array([weather_x, weather_y])
    
    plt.plot(food_vecs[0,], food_vecs[1,], 'ro', color = 'b')
    plt.plot(sports_vecs[0,], sports_vecs[1,], 'ro', color = 'g')
    plt.plot(weather_vecs[0,], weather_vecs[1,], 'ro', color = 'm')
    plt.show()
```

```python
full_pipe('w2v')
```

```python
full_pipe('fastT')
```
