```python
from nltk.corpus import wordnet as wn
wn.synsets('linguistics')
```

```python
print(wn.synset('linguistics.n.01').definition())
print(wn.synset('linguistics.n.02').definition())
```

```python
ling = wn.synset('linguistics.n.01')
try:
    print(ling.hypernyms())
    print("---")
    print(ling.hyponyms())
    print("---")
    print(ling.member_holonyms())
    print("---")
    print(ling.antonyms())
    print("---")
    print(ling.root_hypernyms())
except AttributeError:
    print('No hay atributo')
```

```python
aux = ling.hyponyms()
for _ in aux:
    print(_)
    print(_.definition())
```
