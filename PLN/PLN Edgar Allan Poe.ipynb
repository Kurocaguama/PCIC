```python
import PyPoetryDB, spacy, re, nltk
from PyPoetryDB import PoetryDB

nlp = spacy.load("en_core_web_sm")

poetry_db = PoetryDB.API()
the_raven_raw = poetry_db.get_poem('The Raven')
the_raven = the_raven_raw[0]['lines']
```

```python
the_raven
```

```python
the_raven_text = ''
for _ in the_raven:
    if _ != '':
        the_raven_text = the_raven_text + ' ' + _

the_raven_text = re.sub('[-"!,._;.]|( {2,})', ' ', the_raven_text).lower()
print(the_raven_text[:500])
```

## Types, Tokens, Lemmas, Riqueza

```python
doc = nlp(the_raven_text)
lemmas = [_.lemma_ for _ in doc]
lemmas = [_ for _ in lemmas if _ != ' ']
lemmas = [_ for _ in lemmas if _ != '  ']
lemmas = set(lemmas)
types = the_raven_text.split()
tokens = set(types)
print(len(types), len(tokens), len(lemmas))
```

```python
print(f"Cantidad de palabras totales: {len(types)}. Palabras diferentes: {len(tokens)}. Lemmas: {len(lemmas)}.")
print("La riqueza léxica está definida por la cantidad de palabras diferentes entre la cantidad de palabras totales.")
print(f"Por lo que la riqueza léxia es: {len(tokens)/len(types)}.")
```

## Tags morfosintácticos

```python
the_raven_split = [_ for _ in the_raven if _ != '']
the_raven_split_full = [re.sub('[-"!_]|( {2,})', ' ', _) for _ in the_raven_split]
for _ in the_raven_split_full:
    doc_aux = nlp(_)
    post = [(token, token.pos_) for token in doc_aux if token.pos_ != 'PUNCT']
    post = [_ for _ in post if _[1] != 'SPACE']
    print(post)
    print("---")
```
