```python
proba = float(input('Inserta la probabilidad (dentro del intervalo (0,1) por favor): '))
```

```python
import random
bus = [0 for i in range(40)]
pasajeros = ['P_{}'.format(i) for i in range(40)]
for _ in pasajeros:
    index = pasajeros.index(_)
    rand_val = random.uniform(0,1)
    rand_seat = index
    if rand_val < proba and bus[index] == 0:
        bus[index] = 1
    else:
        while rand_seat == index and bus[rand_seat] != 1:
            rand_seat = random.randint(0,39)
        bus[rand_seat] = 1
        
    print("=====================")
    print("=======INICIO========")
    print('Pasajero: {}'.format(index))
    print('p: {}'.format(rand_val < proba))
    print('Lugar elegido: {}'.format(rand_seat))
    print(bus)
```
