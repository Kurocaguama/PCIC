```python
dropped_index = []
for i in range(len(planets)):
    if np.isnan(planets.iloc[i]['pl_orbper']) or np.isnan(planets.iloc[i]['pl_rade']) or np.isnan(planets.iloc[i]['pl_orbeccen']) or np.isnan(planets.iloc[i]['pl_eqt']) or np.isnan(planets.iloc[i]['st_mass']) or np.isnan(planets.iloc[i]['sy_dist']):
        dropped_index.append(i)
    if type(planets.iloc[i]['st_metratio']) == float:
        dropped_index.append(i)

new_planets = planets.drop(dropped_index)
new_planets = new_planets.reset_index()
new_planets = new_planets.drop(columns = 'index')
new_planets.head()
```

```python
period, radius, excen, temp, mass, grav, dist = [], [], [], [], [], [], []
for i in range(len(new_planets)):
    a = new_planets.iloc[i]
    period.append(a['pl_orbper'])
    radius.append(a['pl_rade'])
    excen.append(a['pl_orbeccen'])
    temp.append(a['pl_eqt'])
    mass.append(a['st_mass'])
    grav.append(a['st_metratio'])
    dist.append(a['sy_dist'])
```

```python
print(period[:5])
print(radius[:5])
print(excen[:5])
print(temp[:5])
print(mass[:5])
print(grav[:5])
print(dist[:5])
```

```python
from sklearn.preprocessing import MinMaxScaler
full_list = [period, radius, excen, temp, mass, dist]
scaled_list = []
for _ in full_list:
    a = MinMaxScaler()
    array = np.array(_)
    array = array.reshape(-1, 1)
    scaled = a.fit_transform(array)
    scaled_list.append(scaled)
print(len(scaled_list))
```

```python
ds_len = len(scaled_list[0])
training_ds = []
for i in range(ds_len):
    data_instance = []
    for _ in scaled_list:
        data_instance.append(_[i])
    aux = np.concatenate(data_instance)
    training_ds.append(aux)

t_ds = np.array(training_ds)
t_ds.shape
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
iso_f = IsolationForest(n_estimators = 30)
lof = LocalOutlierFactor(n_neighbors = 5)

iso_pred = list(iso_f.fit_predict(t_ds))
print('Isolation Forest (15 n_estimators). Cantidad de anomalías: {} | Cantidad de planetas normales: {}'.format(iso_pred.count(-1), iso_pred.count(1)))

lof_pred = list(lof.fit_predict(t_ds))
print('LOF (15 n_neighbors) Cantidad de anomalías: {} | Cantidad de planetas normales: {}'.format(lof_pred.count(-1), lof_pred.count(1)))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import argparse, sys
	
from tensorflow.keras.datasets import mnist
 
(trainX, trainy), (testX, testy) = mnist.load_data()

print('Training data shapes: X=%s, y=%s' % (trainX.shape, trainy.shape))
print('Testing data shapes: X=%s, y=%s' % (testX.shape, testy.shape))
```

```python
import tensorflow

input_data = tensorflow.keras.layers.Input(shape=(6,))
 
#print ("encoder")
encoder = tensorflow.keras.layers.Dense(100)(input_data)
encoder = tensorflow.keras.layers.Activation('relu')(encoder)
 
encoder = tensorflow.keras.layers.Dense(50)(encoder)
encoder = tensorflow.keras.layers.Activation('relu')(encoder)
 
encoder = tensorflow.keras.layers.Dense(25)(encoder)
encoder = tensorflow.keras.layers.Activation('relu')(encoder)
 
encoded = tensorflow.keras.layers.Dense(2)(encoder)

#print ("decoder")
decoder = tensorflow.keras.layers.Dense(25)(encoded)
decoder = tensorflow.keras.layers.Activation('relu')(decoder)
 
decoder = tensorflow.keras.layers.Dense(50)(decoder)
decoder = tensorflow.keras.layers.Activation('relu')(decoder)
 
decoder = tensorflow.keras.layers.Dense(100)(decoder)
decoder = tensorflow.keras.layers.Activation('relu')(decoder)
 
decoded = tensorflow.keras.layers.Dense(6)(decoder)


autoencoder = tensorflow.keras.models.Model(inputs=input_data, outputs=decoded)
autoencoder.compile(loss='mse', optimizer='adam')
autoencoder.summary()
```

```python
from sklearn.model_selection import train_test_split
train_data, test_data = train_test_split(t_ds)
autoencoder.fit(train_data, train_data, epochs=20, batch_size=64, validation_data=(test_data, test_data))

output = autoencoder.predict(t_ds)
output
```

```python
from sklearn.metrics import mean_squared_error
avg_mse = mean_squared_error(t_ds, output)
auto_enc_pred = []

for i in range(len(t_ds)):
    if mean_squared_error(t_ds[i], output[i]) > avg_mse:
        auto_enc_pred.append(-1)
    else:
        auto_enc_pred.append(1)

print('Cantidad de anomalías AutoEncoder: {}. Normales: {}'.format(auto_enc_pred.count(-1), auto_enc_pred.count(1)))
```

```python
%%time
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

tsne = TSNE(n_components = 2, learning_rate = 'auto').fit_transform(t_ds)
pca = PCA(n_components = 2).fit_transform(t_ds)
print(tsne.shape, pca.shape)
```

```python
def get_xy(predictions, proyection, norm):
    anom_index = []
    for i in range(len(predictions)):
        if predictions[i] == -1:
            anom_index.append(i)

    if norm:
        # Anomalía
        planet_list = [(proyection[_][0], proyection[_][1]) if _ in anom_index else None for _ in range(len(proyection))]
    else:
        # Valor normal
        planet_list = [(proyection[_][0], proyection[_][1]) if _ not in anom_index else None for _ in range(len(proyection))]

    planet_list = list(set(planet_list))
    planet_list.remove(None)

    p_x = [planet_list[_][0] for _ in range(len(planet_list))]
    p_y = [planet_list[_][1] for _ in range(len(planet_list))]
    return p_x, p_y
```

```python
ae_norm_tsne_x, ae_norm_tsne_y = get_xy(auto_enc_pred, tsne, False)
ae_anom_tsne_x, ae_anom_tsne_y = get_xy(auto_enc_pred, tsne, True)

fig, ax = plt.subplots()
ax.plot(ae_norm_tsne_x, ae_norm_tsne_y, 'x', color='blue', label='Normal', markersize=6)
ax.plot(ae_anom_tsne_x, ae_anom_tsne_y, '.', color='red', label='Anomalía', markersize=8)
ax.legend(loc='best', fontsize=8)
ax.set_title('Planetas: AE-TSNE')
ax.grid(True, alpha=0.3)
```

```python
ae_norm_pca_x, ae_norm_pca_y = get_xy(auto_enc_pred, pca, False)
ae_anom_pca_x, ae_anom_pca_y = get_xy(auto_enc_pred, pca, True)

fig, ax = plt.subplots()
ax.plot(ae_norm_pca_x, ae_norm_pca_y, 'x', color='blue', label='Normal', markersize=6)
ax.plot(ae_anom_pca_x, ae_anom_pca_y, '.', color='red', label='Anomalía', markersize=8)
ax.legend(loc='best', fontsize=8)
ax.set_title('Planetas: AE-PCA')
ax.grid(True, alpha=0.3)
```
