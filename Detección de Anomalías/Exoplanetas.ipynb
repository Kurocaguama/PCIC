```python
import pandas as pd

planets = pd.read_csv(r'C:\Users\FLopezP\Desktop\PCIC\Tercer Semestre\exoplanets.csv', on_bad_lines = 'skip')
planets.head()
```

```python
import numpy as np
dropped_index = []
for i in range(len(planets)):
    if np.isnan(planets.iloc[i]['pl_orbper']) or np.isnan(planets.iloc[i]['pl_rade']) or np.isnan(planets.iloc[i]['pl_orbeccen']) or np.isnan(planets.iloc[i]['pl_eqt']) or np.isnan(planets.iloc[i]['st_mass']) or np.isnan(planets.iloc[i]['sy_dist']):
        dropped_index.append(i)
    if type(planets.iloc[i]['st_metratio']) == float:
        dropped_index.append(i)

new_planets = planets.drop(dropped_index)
new_planets = new_planets.reset_index()
new_planets = new_planets.drop(columns = 'index')
new_planets.head()
```

```python
period, radius, excen, temp, mass, grav, dist = [], [], [], [], [], [], []
for i in range(len(new_planets)):
    a = new_planets.iloc[i]
    period.append(a['pl_orbper'])
    radius.append(a['pl_rade'])
    excen.append(a['pl_orbeccen'])
    temp.append(a['pl_eqt'])
    mass.append(a['st_mass'])
    grav.append(a['st_metratio'])
    dist.append(a['sy_dist'])
```
